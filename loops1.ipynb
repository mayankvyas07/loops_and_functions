```python
Q1. When to use a for loop and a while loop:

For Loop: Use a for loop when you know the number of iterations in advance or when you want to iterate over a sequence (like a list, tuple, string, etc.). It's particularly useful for iterating over a known range or set of elements.
While Loop: Use a while loop when you want to continue iterating as long as a specific condition is true. It's suitable for situations where you don't know the number of iterations in advance, and you want to keep repeating a block of code until a certain condition is met.
Q2. Python program to print the sum and product of the first 10 natural numbers using for and while loop:

python
Copy code
# Using for loop
sum_for = 0
product_for = 1
for num in range(1, 11):
    sum_for += num
    product_for *= num
print("Sum using for loop:", sum_for)
print("Product using for loop:", product_for)

# Using while loop
sum_while = 0
product_while = 1
num = 1
while num <= 10:
    sum_while += num
    product_while *= num
    num += 1
print("Sum using while loop:", sum_while)
print("Product using while loop:", product_while)
Q3. Python program to compute the electricity bill for a household:

python
Copy code
units_consumed = float(input("Enter the units of electricity consumed: "))

if units_consumed <= 100:
    bill = units_consumed * 4.5
elif units_consumed <= 200:
    bill = 100 * 4.5 + (units_consumed - 100) * 6
elif units_consumed <= 300:
    bill = 100 * 4.5 + 100 * 6 + (units_consumed - 200) * 10
else:
    bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units_consumed - 300) * 20

print("Electricity bill:", bill)
Q4. Python program to calculate the cube of numbers and filter specific conditions:

python
Copy code
numbers = list(range(1, 101))
cubes_divisible_by_4_or_5 = []

for num in numbers:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_divisible_by_4_or_5.append(num)

print("Numbers whose cubes are divisible by 4 or 5:", cubes_divisible_by_4_or_5)
Q5. Python program to count vowels in a given string:

python
Copy code
string = "I want to become a data scientist"
vowels = "aeiouAEIOU"
vowel_count = sum(1 for char in string if char in vowels)
print("Number of vowels in the string:", vowel_count)
```
